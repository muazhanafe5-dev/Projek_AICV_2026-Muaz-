```python
# ============================================================
# PROGRAM: Semakan & Penilaian Purata Markah Pelajar
# Tujuan : Mengira dan menilai prestasi pelajar berdasarkan
#          markah yang dimasukkan oleh pengguna
# ============================================================

# ---- Fungsi untuk menentukan gred berdasarkan markah ----
def dapatkan_gred(markah):
    """
    Fungsi ini menerima markah (0-100) dan mengembalikan gred huruf
    serta status lulus/gagal berdasarkan sistem pemarkahan standard.
    """
    if markah >= 90:
        return "A+", "Cemerlang"
    elif markah >= 80:
        return "A", "Cemerlang"
    elif markah >= 70:
        return "B", "Baik"
    elif markah >= 60:
        return "C", "Memuaskan"
    elif markah >= 50:
        return "D", "Lulus"
    else:
        return "E", "Gagal"


# ---- Fungsi untuk mencetak garis pemisah (hiasan output) ----
def cetak_garis(panjang=50, simbol="="):
    print(simbol * panjang)


# ---- Fungsi utama untuk mendapatkan input markah dari pengguna ----
def masukkan_markah_pelajar():
    """
    Fungsi ini meminta pengguna memasukkan nama pelajar dan
    markah bagi setiap subjek. Ia juga mengesahkan bahawa
    markah yang dimasukkan adalah dalam julat 0 hingga 100.
    """
    cetak_garis()
    print("   SISTEM SEMAKAN PURATA MARKAH PELAJAR")
    cetak_garis()

    # Minta nama pelajar
    nama_pelajar = input("\nMasukkan nama pelajar: ").strip()
    if not nama_pelajar:
        nama_pelajar = "Pelajar"  # Nama lalai jika tidak diisi

    print(f"\nHello, {nama_pelajar}! Mari masukkan markah anda.\n")

    # Minta bilangan subjek
    while True:
        try:
            bilangan_subjek = int(input("Berapa banyak subjek? "))
            if bilangan_subjek <= 0:
                print("  Sila masukkan bilangan subjek yang positif!")
            else:
                break
        except ValueError:
            # Tangkap ralat jika pengguna masukkan bukan nombor
            print("  Sila masukkan nombor sahaja!")

    # Kumpul nama subjek dan markah
    senarai_subjek = []  # Senarai untuk simpan nama subjek
    senarai_markah = []  # Senarai untuk simpan markah subjek

    print()
    for i in range(bilangan_subjek):
        # Minta nama subjek
        nama_subjek = input(f"Nama Subjek {i+1}: ").strip()
        if not nama_subjek:
            nama_subjek = f"Subjek {i+1}"  # Nama lalai jika kosong

        # Minta markah dan pastikan ia dalam julat yang betul
        while True:
            try:
                markah = float(input(f"Markah untuk {nama_subjek} (0-100): "))
                if 0 <= markah <= 100:
                    senarai_subjek.append(nama_subjek)
                    senarai_markah.append(markah)
                    break  # Keluar dari gelung jika markah sah
                else:
                    print("  Markah mesti antara 0 hingga 100!")
            except ValueError:
                print("  Sila masukkan nombor sahaja!")
        print()  # Baris kosong untuk keselesaan bacaan

    return nama_pelajar, senarai_subjek, senarai_markah


# ---- Fungsi untuk mengira statistik markah ----
def kira_statistik(senarai_markah):
    """
    Mengira pelbagai statistik dari senarai markah:
    - Purata (min)
    - Markah tertinggi
    - Markah terendah
    """
    purata = sum(senarai_markah) / len(senarai_markah)  # Jumlah bahagi bilangan
    tertinggi = max(senarai_markah)                      # Markah paling tinggi
    terendah = min(senarai_markah)                       # Markah paling rendah

    return purata, tertinggi, terendah


# ---- Fungsi untuk mencetak laporan penuh pelajar ----
def cetak_laporan(nama_pelajar, senarai_subjek, senarai_markah):
    """
    Mencetak laporan lengkap termasuk:
    - Markah setiap subjek beserta gred
    - Purata keseluruhan
    - Markah tertinggi dan terendah
    - Penilaian akhir prestasi
    """
    cetak_garis()
    print("          LAPORAN PRESTASI PELAJAR")
    cetak_garis()
    print(f"  Nama Pelajar : {nama_pelajar}")
    cetak_garis(simbol="-")

    # Cetak markah dan gred bagi setiap subjek
    print(f"  {'Subjek':<25} {'Markah':>8} {'Gred':>6} {'Status':>12}")
    cetak_garis(simbol="-")

    for subjek, markah in zip(senarai_subjek, senarai_markah):
        gred, status = dapatkan_gred(markah)
        # Format: nama subjek, markah, gred, status (lulus/gagal dll)
        print(f"  {subjek:<25} {markah:>8.1f} {gred:>6} {status:>12}")

    cetak_garis(simbol="-")

    # Kira dan papar statistik keseluruhan
    purata, tertinggi, terendah = kira_statistik(senarai_markah)
    gred_purata, status_purata = dapatkan_gred(purata)

    print(f"\n  Purata Markah  : {purata:.2f}")
    print(f"  Markah Tertinggi: {tertinggi:.1f}")
    print(f"  Markah Terendah : {terendah:.1f}")
    print(f"  Gred Keseluruhan: {gred_purata} ({status_purata})")

    # Mesej motivasi berdasarkan prestasi
    cetak_garis(simbol="-")
    if purata >= 80:
        print("  Tahniah! Prestasi anda sangat cemerlang!")
    elif purata >= 60:
        print("  Bagus! Teruskan usaha anda untuk lebih baik.")
    elif purata >= 50:
        print("  Anda lulus. Tingkatkan lagi usaha anda!")
    else:
        print("  Jangan berputus asa. Berusaha lebih keras lagi!")

    cetak_garis()


# ---- Bahagian utama program (entry point) ----
def main():
    """
    Fungsi utama yang menjalankan keseluruhan program.
    Ia juga membenarkan pengguna semak pelajar lain
    tanpa perlu jalankan semula program.
    """
    while True:
        # Dapatkan maklumat dan markah pelajar
        nama_pelajar, senarai_subjek, senarai_markah = masukkan_markah_pelajar()

        # Cetak laporan prestasi
        print()
        cetak_laporan(nama_pelajar, senarai_subjek, senarai_markah)

        # Tanya sama ada ingin semak pelajar lain
        print()
        ulang = input("Semak pelajar lain? (y/tidak): ").strip().lower()
        if ulang not in ("y", "ya", "yes"):
            print("\nTerima kasih. Program tamat.")
            break
        print()  # Baris kosong sebelum bermula semula


# ---- Titik permulaan program ----
# Baris ini memastikan 'main()' hanya dipanggil apabila
# fail ini dijalankan terus, bukan apabila di-import
if __name__ == "__main__":
    main()
```

```python
# ================================================================
# PROGRAM  : Sistem Pengiraan Indeks Jisim Badan (BMI)
# Bahasa   : Python 3
# Tujuan   : Mengira BMI pengguna dan menilai kategori berat badan
# ================================================================


# ----------------------------------------------------------------
# FUNGSI 1: cetak_garis()
# Tujuan  : Mencetak garis pemisah supaya output kelihatan kemas
# ----------------------------------------------------------------
def cetak_garis(panjang=55, simbol="="):
    # 'simbol * panjang' = ulang simbol sebanyak 'panjang' kali
    # contoh: "=" * 5  -->  "====="
    print(simbol * panjang)


# ----------------------------------------------------------------
# FUNGSI 2: dapatkan_kategori_bmi()
# Tujuan  : Tentukan kategori BMI berdasarkan nilai yang dikira
# Input   : nilai BMI (nombor perpuluhan)
# Output  : (kategori, nasihat, simbol emoji)
# ----------------------------------------------------------------
def dapatkan_kategori_bmi(bmi):

    # Semak sama ada BMI dalam julat "Kurus Teruk"
    if bmi < 16.0:
        return "Kurus Teruk (Underweight)", "Anda perlu mendapatkan rawatan perubatan segera.", "⚠️"

    # Semak sama ada BMI dalam julat "Kurus"
    elif bmi < 18.5:
        return "Kurus (Underweight)", "Tingkatkan pengambilan kalori dan jumpa doktor.", "🔵"

    # Semak sama ada BMI dalam julat "Normal" (sihat)
    elif bmi < 25.0:
        return "Normal (Sihat)", "Tahniah! Kekalkan gaya hidup sihat anda.", "✅"

    # Semak sama ada BMI dalam julat "Berat Berlebihan"
    elif bmi < 30.0:
        return "Berat Berlebihan (Overweight)", "Kurangkan makanan berlemak dan aktifkan senaman.", "🟡"

    # Semak sama ada BMI dalam julat "Obes Tahap 1"
    elif bmi < 35.0:
        return "Obes Tahap 1", "Disyorkan berjumpa doktor untuk program diet.", "🟠"

    # Semak sama ada BMI dalam julat "Obes Tahap 2"
    elif bmi < 40.0:
        return "Obes Tahap 2", "Segera dapatkan nasihat perubatan profesional.", "🔴"

    # Jika semua syarat di atas tidak sepadan, bermakna BMI >= 40
    else:
        return "Obes Teruk (Tahap 3)", "Perlu rawatan perubatan segera. Sila jumpa doktor.", "🚨"


# ----------------------------------------------------------------
# FUNGSI 3: kira_berat_ideal()
# Tujuan  : Mengira julat berat badan ideal berdasarkan ketinggian
#           menggunakan formula BMI Normal (18.5 - 24.9)
# Input   : ketinggian dalam meter
# Output  : (berat_min, berat_max) dalam kg
# ----------------------------------------------------------------
def kira_berat_ideal(tinggi_m):

    # Formula: Berat = BMI × Tinggi²
    # Berat minimum = BMI paling rendah normal (18.5) × tinggi²
    berat_min = 18.5 * (tinggi_m ** 2)

    # Berat maximum = BMI paling tinggi normal (24.9) × tinggi²
    berat_max = 24.9 * (tinggi_m ** 2)

    # Kembalikan dua nilai sekaligus (tuple)
    return berat_min, berat_max


# ----------------------------------------------------------------
# FUNGSI 4: dapatkan_input_pengguna()
# Tujuan  : Meminta dan mengesahkan input dari pengguna
#           (nama, jantina, umur, berat, ketinggian)
# Output  : Semua data pengguna yang telah disahkan
# ----------------------------------------------------------------
def dapatkan_input_pengguna():

    # Cetak tajuk program
    cetak_garis()
    print("      SISTEM PENGIRAAN INDEKS JISIM BADAN (BMI)")
    cetak_garis()

    # ---- Input Nama ----
    # .strip() = buang ruang kosong di depan dan belakang teks
    nama = input("\nMasukkan nama anda : ").strip()

    # Jika nama kosong, gunakan nama lalai
    if not nama:
        nama = "Pengguna"

    # ---- Input Jantina ----
    while True:                                      # Gelung sehingga input sah
        jantina = input("Jantina (L/P)      : ").strip().upper()  # .upper() = tukar ke huruf besar
        if jantina in ("L", "P"):                   # Semak sama ada L atau P sahaja
            break                                    # Keluar gelung jika sah
        print("  Input tidak sah! Sila masukkan L (Lelaki) atau P (Perempuan).")

    # ---- Input Umur ----
    while True:
        try:
            umur = int(input("Umur (tahun)       : "))   # Tukar input kepada integer
            if 1 <= umur <= 120:                          # Pastikan umur munasabah
                break
            print("  Sila masukkan umur antara 1 hingga 120 tahun.")
        except ValueError:                               # Tangkap ralat jika bukan nombor
            print("  Sila masukkan nombor sahaja!")

    # ---- Input Berat Badan ----
    while True:
        try:
            berat_kg = float(input("Berat badan (kg)   : "))   # float() = terima nombor perpuluhan
            if 1.0 <= berat_kg <= 500.0:                        # Had berat yang munasabah
                break
            print("  Sila masukkan berat antara 1 hingga 500 kg.")
        except ValueError:
            print("  Sila masukkan nombor sahaja!")

    # ---- Input Ketinggian ----
    # Tanya unit yang digunakan (cm atau meter)
    while True:
        unit = input("Unit ketinggian (cm/m): ").strip().lower()  # .lower() = tukar ke huruf kecil
        if unit in ("cm", "m"):
            break
        print("  Sila masukkan 'cm' atau 'm' sahaja.")

    while True:
        try:
            tinggi_input = float(input(f"Ketinggian ({unit})  : "))

            # Tukar ke meter bergantung pada unit yang dipilih
            if unit == "cm":
                tinggi_m = tinggi_input / 100   # Bahagi 100 untuk tukar cm ke meter
            else:
                tinggi_m = tinggi_input         # Sudah dalam meter, guna terus

            # Pastikan ketinggian munasabah (0.5m - 2.5m)
            if 0.5 <= tinggi_m <= 2.5:
                break
            print("  Sila masukkan ketinggian yang munasabah (50cm - 250cm).")
        except ValueError:
            print("  Sila masukkan nombor sahaja!")

    # Kembalikan semua data yang telah dikumpul
    return nama, jantina, umur, berat_kg, tinggi_m


# ----------------------------------------------------------------
# FUNGSI 5: cetak_laporan_bmi()
# Tujuan  : Memaparkan laporan BMI yang lengkap dan terperinci
# Input   : Semua maklumat pengguna + nilai BMI yang telah dikira
# ----------------------------------------------------------------
def cetak_laporan_bmi(nama, jantina, umur, berat_kg, tinggi_m, bmi):

    # Dapatkan kategori dan nasihat berdasarkan nilai BMI
    kategori, nasihat, emoji = dapatkan_kategori_bmi(bmi)

    # Dapatkan julat berat ideal untuk ketinggian ini
    berat_min, berat_max = kira_berat_ideal(tinggi_m)

    # Tentukan label jantina penuh untuk paparan
    label_jantina = "Lelaki" if jantina == "L" else "Perempuan"

    # Kira berapa kg perlu tambah atau kurang untuk mencapai berat ideal
    if berat_kg < berat_min:
        # Berat kurang dari had minimum — perlu tambah berat
        perbezaan = berat_min - berat_kg
        mesej_perbezaan = f"Perlu tambah {perbezaan:.1f} kg untuk mencapai berat minimum ideal."
    elif berat_kg > berat_max:
        # Berat melebihi had maksimum — perlu kurang berat
        perbezaan = berat_kg - berat_max
        mesej_perbezaan = f"Perlu kurang {perbezaan:.1f} kg untuk mencapai berat maksimum ideal."
    else:
        # Berat sudah dalam julat ideal
        mesej_perbezaan = "Berat badan anda sudah dalam julat ideal!"

    # ---- Cetak laporan ----
    print()
    cetak_garis()
    print("              LAPORAN BMI ANDA")
    cetak_garis()

    # Maklumat peribadi
    print(f"  Nama          : {nama}")
    print(f"  Jantina       : {label_jantina}")
    print(f"  Umur          : {umur} tahun")

    cetak_garis(simbol="-")

    # Maklumat ukuran badan
    print(f"  Berat Badan   : {berat_kg:.1f} kg")
    print(f"  Ketinggian    : {tinggi_m * 100:.1f} cm  ({tinggi_m:.2f} m)")

    cetak_garis(simbol="-")

    # Keputusan BMI — format ke 2 titik perpuluhan
    print(f"  Nilai BMI     : {bmi:.2f}")
    print(f"  Kategori      : {emoji}  {kategori}")

    cetak_garis(simbol="-")

    # Maklumat berat ideal
    print(f"  Berat Ideal   : {berat_min:.1f} kg  —  {berat_max:.1f} kg")
    print(f"  Status        : {mesej_perbezaan}")

    cetak_garis(simbol="-")

    # Nasihat kesihatan
    print(f"  Nasihat       : {nasihat}")

    cetak_garis()


# ----------------------------------------------------------------
# FUNGSI 6: papar_carta_bmi()
# Tujuan  : Memaparkan carta rujukan kategori BMI untuk pengguna
# ----------------------------------------------------------------
def papar_carta_bmi():
    print("\n  CARTA RUJUKAN BMI:")
    cetak_garis(simbol="-")

    # Setiap baris mengandungi: julat BMI, kategori, simbol
    carta = [
        ("< 16.0",      "Kurus Teruk",            "⚠️ "),
        ("16.0 - 18.4", "Kurus (Underweight)",     "🔵"),
        ("18.5 - 24.9", "Normal (Sihat)",           "✅"),
        ("25.0 - 29.9", "Berat Berlebihan",         "🟡"),
        ("30.0 - 34.9", "Obes Tahap 1",             "🟠"),
        ("35.0 - 39.9", "Obes Tahap 2",             "🔴"),
        (">= 40.0",     "Obes Teruk (Tahap 3)",     "🚨"),
    ]

    # Cetak setiap baris carta menggunakan gelung
    for julat, kategori, simbol in carta:
        # Format kolum supaya tersusun kemas
        print(f"  {julat:<15}  {simbol}  {kategori}")

    cetak_garis(simbol="-")


# ----------------------------------------------------------------
# FUNGSI 7: main()
# Tujuan  : Fungsi utama — menghubungkan semua fungsi di atas
#           dan mengawal aliran keseluruhan program
# ----------------------------------------------------------------
def main():

    while True:   # Gelung utama — program terus berjalan sehingga pengguna keluar

        # Langkah 1: Dapatkan semua maklumat dari pengguna
        nama, jantina, umur, berat_kg, tinggi_m = dapatkan_input_pengguna()

        # Langkah 2: Kira nilai BMI menggunakan formula standard WHO
        # Formula BMI = Berat (kg) ÷ Tinggi² (m²)
        # Contoh: 70kg ÷ (1.70m × 1.70m) = 70 ÷ 2.89 = 24.22
        bmi = berat_kg / (tinggi_m ** 2)

        # Langkah 3: Cetak laporan lengkap BMI pengguna
        cetak_laporan_bmi(nama, jantina, umur, berat_kg, tinggi_m, bmi)

        # Langkah 4: Tanya sama ada pengguna ingin lihat carta rujukan
        tanya_carta = input("\nIngin lihat carta rujukan BMI? (y/t): ").strip().lower()
        if tanya_carta in ("y", "ya"):
            papar_carta_bmi()   # Papar carta rujukan jika pengguna mahu

        # Langkah 5: Tanya sama ada ingin kira BMI orang lain
        print()
        ulang = input("Kira BMI orang lain? (y/t): ").strip().lower()
        if ulang not in ("y", "ya"):
            # Jika tidak, cetak mesej perpisahan dan tamatkan program
            print("\n  Jaga kesihatan anda! Program tamat.")
            cetak_garis()
            break   # Keluar dari gelung utama, program berakhir


# ----------------------------------------------------------------
# TITIK PERMULAAN PROGRAM
# '__name__ == "__main__"' bermaksud: jalankan main() hanya
# apabila fail ini dijalankan terus (bukan di-import dari fail lain)
# ----------------------------------------------------------------
if __name__ == "__main__":
    main()
```
