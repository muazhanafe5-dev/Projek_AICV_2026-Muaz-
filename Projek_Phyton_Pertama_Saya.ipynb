```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Guna cara Numpy
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots() # Buat kanvas
ax.plot(x, y1, color='blue', label='Sin')
ax.plot(x, y2, color='red', label='cos')

ax.set_title("Perbandingan Sin & Cos")
ax.legend()
plt.show()
```
